```python
import pandas as pd
import numpy as np
# Load the dataset
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

print(" Experiment 1 ")
print(df.head())

num_cols = df.select_dtypes(include=np.number)
cat_cols = df.select_dtypes(exclude=np.number)

print("\nNumerical Columns:\n", num_cols.columns.tolist())
print("\nCategorical Columns:\n", cat_cols.columns.tolist())

print("\nMean:\n", num_cols.mean())
print("\nMedian:\n", num_cols.median())
print("\nMode:\n", num_cols.mode().iloc[0])

print("\nMinimum:\n", num_cols.min())
print("\nMaximum:\n", num_cols.max())
print("\nSum:\n", num_cols.sum())
print("\nVariance:\n", num_cols.var())
print("\nStandard Deviation:\n", num_cols.std())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("StudentsPerformance.csv")

# Select numerical columns
numerical_cols = df.select_dtypes(include='number')

# Compute quartiles
Q1 = numerical_cols.quantile(0.25)
Q2 = numerical_cols.quantile(0.50)  # Median
Q3 = numerical_cols.quantile(0.75)

print("First Quartile (Q1):\n", Q1)
print("\nSecond Quartile (Q2 - Median):\n", Q2)
print("\nThird Quartile (Q3):\n", Q3)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("StudentsPerformance.csv")

# Select numerical columns
numerical_cols = df.select_dtypes(include='number')

# Compute correlation matrix
correlation_matrix = numerical_cols.corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Compute covariance matrix
covariance_matrix = numerical_cols.cov()
print("\nCovariance Matrix:\n", covariance_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Select numerical columns
numerical_cols = df.select_dtypes(include='number')

# ---------- Histograms ----------
for col in numerical_cols.columns:
    plt.figure(figsize=(7, 4))
    plt.hist(
        df[col],
        bins=15,                 # fewer bins for smoother shape
        edgecolor='black',       # clear separation
        linewidth=0.8
    )
    plt.xlabel(col, fontsize=11)
    plt.ylabel("Frequency", fontsize=11)
    plt.title(f"Distribution of {col}", fontsize=13)
    plt.tight_layout()
    plt.show()

# ---------- Boxplots ----------
for col in numerical_cols.columns:
    plt.figure(figsize=(4, 5))
    plt.boxplot(
        df[col],
        widths=0.5
    )
    plt.ylabel(col, fontsize=11)
    plt.title(f"Boxplot of {col}", fontsize=13)
    plt.grid(axis='y', linestyle='--', alpha=0.6)  # subtle grid only on y-axis
    plt.tight_layout()
    plt.show()
```
